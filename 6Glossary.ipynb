# Glossary

## Acronyms

- **HEP:** High Energy Physics
- **MC:** [Monte Carlo](#mc)
- **PID:** Particle IDentifier
- **CC:** Charged-current

## Terms

- <a id='mc' /> **Monte Carlo:** Shorthand for _simulated experiment_, named after the city famous for it's casinos because simulations make heavy use of random number generation (rolling the die).
  + <a id='mc_truth' /> **Truth:** Truth information is information that we can access about the simulation that is not accessible in a real experiment. For example, the energy of a neutrino before it interacts is unknowable because our particle detectors are not sensitive to neutrinos travelling through them and must be _reconstructed_ from observable particles produced when a neutrino interacts with some matter in our detector.

- <a id='event' /> **Event:** Usually used to refer to a single particle interaction of interest, for us it corresponds to a neutrino interacting with the matter in one of our detectors.
- <a id='pid' /> **Particle IDentifier:** A unique number corresponding to a particle species of particle. Some useful ones for this project:
  + 11: Electron ($\mathrm{e}^{-}$)
  + 12: Electron Neutrino ($\nu_\mathrm{e}$)
  + 13: Muon ($\mu$)
  + 14: Muon Neutrino ($\nu_\mu$)

- <a id='cc' /> **Charged Current:** An interaction that involves a transfer of electric charge from one particle system to another. For this project it means interactions in which a neutrino comes in and a charged lepton ($\mu$ or $\mathrm{e}^{-}$) leaves. This is as opposed to a _Neutral Current_ interaction where a neutrino comes in and a neutrino of the same [flavor](#flavor) leaves.
